## 2. Change learning rate

$C(w) = (w+3)^2 /2$  

$C'(w) = w+3$

```python
w = 0           # initial
a = 0.5      # learning rate
w_list=[]

for i in range(30):
    g = w + 3       # gradient
    w = w - a*g
    c = (w+3)**2/2 # cost function
    w_list.append(w)
    
plt.plot(w_list, "o")
plt.show()
```

## 3. Find optimal value with gradient descent

```python

x = np.array([1, 2, 3])
y = np.array([3.1, 5.9, 8.9])
w = 0
a = 0.05      # 학습률

H = w*x
# print(H)
for i in range(10):
    g = 2*sum(x*(H-y))/len(y)
    w = w - a*g
    H = w*x
    c = sum((H-y)*(H-y))/len(y)
    print("w: ", w, "loss: ", c)
    
print("prediction:", w*4)    
```
